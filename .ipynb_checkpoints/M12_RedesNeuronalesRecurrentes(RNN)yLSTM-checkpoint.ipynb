**Paso 2: Importación de Datos**

Dado que tienes poca RAM, vamos a usar un conjunto de datos pequeño integrado dentro del código. Usaremos datos generados para la predicción de una serie temporal simple.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generamos datos simulados para una serie temporal
data_length = 1000
time_steps = np.arange(data_length)
data_series = np.sin(time_steps * 0.02) + np.random.normal(0, 0.1, size=data_length)

# Creamos un DataFrame
df = pd.DataFrame({'Time': time_steps, 'Value': data_series})

# Mostramos los primeros datos
print(df.head())

# Visualizamos la serie temporal
plt.plot(df['Time'], df['Value'])
plt.title('Serie temporal simulada')
plt.show()
```

**Paso 3: Preprocesamiento de Datos**

El LSTM trabaja mejor con datos normalizados y secuenciales. Aquí, escalamos los valores entre 0 y 1 usando MinMaxScaler.

```python
# Escalar los datos para que estén entre 0 y 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Value'].values.reshape(-1, 1))

# Función para crear secuencias de datos (X) y sus correspondientes etiquetas (y)
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

# Definimos la longitud de la secuencia
sequence_length = 50

# Crear secuencias y etiquetas
X, y = create_sequences(scaled_data, sequence_length)

# Liberamos memoria innecesaria
del scaled_data
```

**Paso 4: División en Conjuntos de Entrenamiento y Prueba**

Vamos a dividir los datos en conjuntos de entrenamiento (80%) y prueba (20%).

```python
# Dividimos en conjuntos de entrenamiento y prueba
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"Tamaño del conjunto de entrenamiento: {X_train.shape}, {y_train.shape}")
print(f"Tamaño del conjunto de prueba: {X_test.shape}, {y_test.shape}")
```

**Paso 5: Construcción del Modelo LSTM**

Para ahorrar RAM, limitaremos el tamaño del modelo y usaremos un modelo LSTM básico.

```python
# Construimos el modelo LSTM
model = Sequential()

# Capa LSTM con 50 unidades
model.add(LSTM(50, return_sequences=False, input_shape=(X_train.shape[1], 1)))

# Capa de salida densa (1 neurona)
model.add(Dense(1))

# Compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Mostramos la estructura del modelo
model.summary()
```

**Paso 6: Entrenamiento del Modelo**

Entrenaremos el modelo con un número reducido de épocas para optimizar la memoria.

```python
# Entrenar el modelo (usamos solo 5 épocas para optimizar tiempo y memoria)
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

# Visualizamos la evolución de la pérdida
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('Evolución de la pérdida')
plt.legend()
plt.show()
```

**Paso 7: Evaluación del Modelo**

Una vez entrenado el modelo, podemos hacer predicciones sobre el conjunto de prueba y desescalar los resultados.

```python
# Hacemos predicciones sobre el conjunto de prueba
predictions = model.predict(X_test)

# Desescalamos los valores para obtener los originales
predictions_rescaled = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Visualizamos los resultados
plt.plot(y_test_rescaled, label='Real')
plt.plot(predictions_rescaled, label='Predicción')
plt.title('Predicción vs Real')
plt.legend()
plt.show()
```

**Paso 8: Predicción de Valores Futuros**

Podemos usar el modelo entrenado para predecir valores futuros.

```python
# Predicción de los siguientes 10 pasos en el futuro
future_steps = 10
last_sequence = X_test[-1]  # Última secuencia del conjunto de prueba
future_predictions = []

for _ in range(future_steps):
    pred = model.predict(last_sequence.reshape(1, sequence_length, 1))
    future_predictions.append(pred[0][0])
    last_sequence = np.append(last_sequence[1:], pred, axis=0)

# Desescalamos los valores futuros
future_predictions_rescaled = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Visualizamos los valores futuros
plt.plot(future_predictions_rescaled, label='Predicciones futuras')
plt.title('Predicción de los siguientes 10 pasos')
plt.show()
```
