<H1>Módulo 5: Clasificación con Regresión Logística </H1>

### 1. Introducción a la Regresión Logística

La **regresión logística** es un modelo de clasificación utilizado para predecir la probabilidad de que una observación pertenezca a una clase binaria (0 o 1). A diferencia de la regresión lineal, que predice valores continuos, la regresión logística predice probabilidades que se mapean a una clase binaria mediante la función sigmoide:

$
P(Y=1 | \mathbf{X}) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p)}}
$

donde:
- $P(Y=1 | \mathbf{X})$ es la probabilidad de que la clase sea 1 dado el vector de características $\mathbf{X}$.
- $\beta_0$, $\beta_1$, $\beta_2$, $\ldots$, $\beta_p$ son los coeficientes del modelo.
- $e$ es la base del logaritmo natural.

### 2. Estimación de los Coeficientes $\beta$

Los coeficientes $\beta$ se estiman utilizando métodos de máxima verosimilitud. 

**Métodos de Verosimilitud**:

La función de verosimilitud para el modelo de regresión logística es:

$
L(\beta) = \prod_{i=1}^{n} \left[ P(Y_i=1 | \mathbf{X_i})^{y_i} \cdot \left(1 - P(Y_i=1 | \mathbf{X_i})\right)^{1 - y_i} \right]
$

donde $y_i$ es la variable dependiente (0 o 1) para la observación $i$.

Para encontrar los valores óptimos de $\beta$, se maximiza la función de log-verosimilitud:

$
\text{Log-Likelihood} = \sum_{i=1}^{n} \left[ y_i \log P(Y_i=1 | \mathbf{X_i}) + (1 - y_i) \log \left(1 - P(Y_i=1 | \mathbf{X_i})\right) \right]
$

### 3. Métricas de Clasificación

#### **3.1. Accuracy (Exactitud)**

La **exactitud** mide la proporción de predicciones correctas entre todas las predicciones realizadas. Se calcula como:

$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$

donde:
- $TP$ es el número de verdaderos positivos.
- $TN$ es el número de verdaderos negativos.
- $FP$ es el número de falsos positivos.
- $FN$ es el número de falsos negativos.

#### **3.2. Precision (Precisión)**

La **precisión** mide la proporción de verdaderos positivos entre todas las instancias clasificadas como positivas:

$
\text{Precision} = \frac{TP}{TP + FP}
$

#### **3.3. Recall (Exhaustividad o Sensibilidad)**

El **recall** mide la proporción de verdaderos positivos entre todas las instancias que realmente son positivas:

$
\text{Recall} = \frac{TP}{TP + FN}
$

#### **3.4. F1-Score**

El **F1-Score** combina precisión y recall para proporcionar una medida única del rendimiento del modelo:

$
\text{F1-Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$

#### **3.5. ROC-AUC**

El **Área bajo la Curva ROC (ROC-AUC)** mide la capacidad del modelo para distinguir entre las clases a diferentes umbrales. La curva ROC traza la tasa de verdaderos positivos frente a la tasa de falsos positivos a diferentes umbrales.

La función ROC en un umbral $t$ se define como:

$
\text{ROC}(t) = \frac{\text{TPR}(t)}{\text{FPR}(t)}
$

donde:
- $\text{TPR}(t)$ (True Positive Rate o Tasa de Verdaderos Positivos) se calcula como:

$
\text{TPR}(t) = \frac{TP(t)}{TP(t) + FN(t)}
$

- $\text{FPR}(t)$ (False Positive Rate o Tasa de Falsos Positivos) se calcula como:

$
\text{FPR}(t) = \frac{FP(t)}{FP(t) + TN(t)}
$

El AUC se calcula como el área bajo la curva ROC:

$
\text{AUC} = \int_{0}^{1} \text{ROC}(t) \, dt
$

### 4. Relación entre Log-Likelihood, Probabilidades Predichas y Métricas

- **Accuracy**: Un mayor log-likelihood indica que las probabilidades predichas están más cerca de los valores reales de la clase, lo que puede llevar a una mayor exactitud. La probabilidad $P$ se ajusta para maximizar el log-likelihood, lo cual mejora la exactitud.

- **Precision y Recall**: La precisión y el recall se calculan a partir de las predicciones binarizadas basadas en el umbral de probabilidad $P$. Un log-likelihood más alto generalmente indica que las probabilidades predichas son más precisas, lo que puede llevar a una mejor precisión y recall. Ajustar el umbral $t$ afectará estos valores.

- **F1-Score**: El F1-Score es una medida combinada de precisión y recall, y se verá afectado por el ajuste del umbral. Un log-likelihood más alto sugiere que el modelo ajusta mejor las probabilidades, lo que puede mejorar el F1-Score.

- **ROC-AUC**: El AUC refleja la capacidad del modelo para distinguir entre las clases a diferentes umbrales. Un log-likelihood mayor generalmente se asocia con un mejor ajuste del modelo, lo que puede resultar en una curva ROC más alta y, por lo tanto, un AUC mayor. La función ROC se basa en las tasas de verdaderos y falsos positivos, que se ven afectadas por cómo las probabilidades predichas se comparan con el umbral.

Proyecto: Clasificación de supervivencia en el Titanic.
Utilizar el dataset del Titanic y predecir la supervivencia de los pasajeros implementando un modelo de regresión logística.

**Conceptos Clave**

**Regresión Logística:** Es un modelo estadístico que se utiliza para la clasificación binaria. En lugar de predecir una variable continua, como en la regresión lineal, la regresión logística predice la probabilidad de que una instancia pertenezca a una de las dos clases posibles (por ejemplo, sobrevivió o no sobrevivió en el contexto del Titanic).

**Clasificación Binaria:** Es un tipo de tarea de clasificación donde solo hay dos posibles categorías para cada instancia. En el proyecto del Titanic, las dos categorías son "Sobrevivió" o "No sobrevivió".

**Métricas de Clasificación:**

Accuracy (Precisión): Es la proporción de predicciones correctas sobre el total de instancias. 

Se calcula como 
$(
(TP+TN)/(TP+TN+FP+FN)$, 

donde TP es Verdaderos Positivos, TN es Verdaderos Negativos, FP es Falsos Positivos y FN es Falsos Negativos.



**Precision (Precisión):** Es la proporción de verdaderos positivos sobre el total de predicciones positivas. 

Se calcula como 
$
TP/(TP+FP)$.

Mide la exactitud de las predicciones positivas.

**Recall (Exhaustividad):** Es la proporción de verdaderos positivos sobre el total de casos positivos reales. Se calcula como 
$
TP/(TP+FN)$. Mide la capacidad del modelo para encontrar todos los casos positivos.

**F1-Score**: Es la media armónica entre precisión y exhaustividad. Se calcula como 
$
2×(Precision×Recall)/(Precision+Recall)$. Es útil cuando hay un desequilibrio entre las clases.

**ROC-AUC (Área bajo la curva ROC)**: Es una métrica que evalúa el rendimiento del modelo en diferentes umbrales de clasificación. La curva ROC (Receiver Operating Characteristic) muestra la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos.

**Proyecto: Clasificación de Supervivencia en el Titanic**

**Paso 1: Importar Librerías y Cargar Datos**
Primero, importaremos las librerías necesarias y cargaremos el dataset del Titanic. Utilizaremos el dataset de Kaggle.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve


# Cargar el dataset del Titanic
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
data = pd.read_csv(url)
```

```python
# Mostrar las primeras filas del dataset y las columnas
print(data.head())
print(data.columns)
```

**3. Preprocesar los Datos**
 
Elimina columnas no útiles, convierte variables categóricas en variables dummy, y maneja los valores faltantes si es necesario.

```python
# Eliminar la columna 'Name' que no es necesaria para la clasificación
data = data.drop(columns=['Name'])

# Convertir la columna 'Sex' en variables dummy
data = pd.get_dummies(data, columns=['Sex'], drop_first=True)

# Verificar datos procesados
print(data.head())
```

**4. Definir Variables Predictoras y Variable Objetivo**

Separa las características (X) de la variable objetivo (y).

```python
# Definir variables predictoras (X) y variable objetivo (y)
X = data.drop(columns=['Survived'])
y = data['Survived']
```

**5. Dividir el Dataset**

Divide los datos en conjuntos de entrenamiento y prueba.

```python
# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**6. Escalar las Características**

Escala las características para que tengan el mismo rango, lo que ayuda a mejorar el rendimiento del modelo.

```python
# Escalar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**7. Entrenar el Modelo**

Inicializa y entrena el modelo de regresión logística.

```python
# Inicializar y entrenar el modelo de regresión logística
model = LogisticRegression()
model.fit(X_train, y_train)
```

**8. Realizar Predicciones**

Usa el modelo para hacer predicciones sobre el conjunto de prueba.

```python
# Realizar predicciones
y_pred = model.predict(X_test)
```

**9. Evaluar el Modelo**

Calcula y muestra las métricas de rendimiento del modelo.

```python
# Calcular métricas de clasificación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
```

```python
# Mostrar métricas
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"ROC-AUC: {roc_auc:.2f}")
```

una explicación breve de cada métrica:

1. **Accuracy (Exactitud) - 0.77**: 
   - **Definición**: Proporción de predicciones correctas (tanto positivas como negativas) sobre el total de predicciones.
   - **Interpretación**: El modelo acertó el 77% de las veces. Es una medida general de desempeño.

2. **Precision (Precisión) - 0.75**: 
   - **Definición**: Proporción de verdaderos positivos sobre el total de predicciones positivas realizadas por el modelo.
   - **Interpretación**: De todas las veces que el modelo predijo que un pasajero sobreviviría, el 75% de las veces la predicción fue correcta.

3. **Recall (Sensibilidad) - 0.57**: 
   - **Definición**: Proporción de verdaderos positivos sobre el total de casos realmente positivos en el conjunto de datos.
   - **Interpretación**: El modelo identificó correctamente el 57% de los pasajeros que realmente sobrevivieron. 

4. **F1-Score - 0.65**: 
   - **Definición**: Media armónica de la precisión y el recall, proporcionando una única métrica que equilibra ambos.
   - **Interpretación**: Un valor de 0.65 indica un equilibrio entre precisión y recall, siendo un buen resumen del rendimiento general en problemas de clasificación.

5. **ROC-AUC - 0.73**: 
   - **Definición**: Área bajo la curva ROC, que mide la capacidad del modelo para distinguir entre las clases positivas y negativas.
   - **Interpretación**: Un valor de 0.73 indica que el modelo tiene un buen desempeño en la discriminación entre las clases. El AUC va de 0.5 (sin capacidad de discriminación) a 1 (capacidad de discriminación perfecta).

**10. Graficar la Curva ROC**

Grafica la curva ROC para visualizar el desempeño del modelo en diferentes umbrales.

```python
# Graficar la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
